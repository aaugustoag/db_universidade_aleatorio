```python
import random
import pandas as pd
```

```python
def geraMaterias(departamento, abreviacaoDepartamento, materiaPorDepartamento):
  materia = []
  for i in range(len(abreviacaoDepartamento)):
    materia.append([departamento[i]])
    for j in range(materiaPorDepartamento[i]):
      materia[i].append(abreviacaoDepartamento[i] + str(j+1).zfill(3))
    if len(materia[i]) < 16:
      for j in range(16-len(materia[i])):
        materia[i].append('None')

  return materia
```

```python
departamento = ['Matematica', 'Fisica', 'Humanas', 'Computacao', 'Administrativo', 'Linguagens']
abreviacaoDepartamento = ['mat', 'fis', 'hum', 'comp', 'adm', 'ling']
materiaPorDepartamento = [4, 3, 2, 11, 0, 3]
materia = geraMaterias(departamento, abreviacaoDepartamento, materiaPorDepartamento)
dfMateria=pd.DataFrame(materia, columns=['Departamento', 'Materia1', 'Materia2', 'Materia3',
                                             'Materia4', 'Materia5', 'Materia6', 'Materia7',
                                             'Materia8', 'Materia9', 'Materia10', 'Materia11',
                                             'Materia12', 'Materia13', 'Materia14', 'Materia15'])
dfMateria.set_index('Departamento', inplace=True)
dfMateria
```

```python
def gerarMatricula(qtdMatricula, departamento, cargo):
  matricula = []
  for i in range(qtdMatricula):
    matricula.append([str(i+1).zfill(6), random.choice(cargo), random.choice(departamento)])
  return matricula
```

```python
cargo = ['Docente', 'Dicente', 'Tecnico', 'Administrativo']
matricula = gerarMatricula(1000000, departamento, cargo) #esse 10 é a quantidade de dados
dfMatricula=pd.DataFrame(matricula, columns=['Matricula', 'Cargo', 'Departamento'])
dfMatricula.set_index('Matricula', inplace=True)
dfMatricula
```

```python
def materiaAleatoria (materia):
  dpRandom=random.randrange(len(materia))
  turma=random.choice(materia[dpRandom][1:])
  return turma

def gerarHorario(matricula, materia):
  horario = []
  for matr in matricula:
    if matr[1]=='Dicente':
      horarioPorPeriodo = []
      while len(horarioPorPeriodo) < 6:
        mat = materiaAleatoria (materia)
        while mat in horarioPorPeriodo:
          mat = materiaAleatoria (materia)
        if mat != 'None':
          horarioPorPeriodo.append(mat)
      horario.append([matr[0]] + horarioPorPeriodo)

  return horario
```

```python
horario=gerarHorario(matricula, materia)
dfHorario=pd.DataFrame(horario, columns=['Matricula', 'M1', 'M2', 'M3', 'T1', 'T2', 'T3'])
dfHorario.set_index('Matricula', inplace=True)
dfHorario
```

# Gerar CSVs

```python

dfMateria.to_csv('Materia.csv')
dfMatricula.to_csv('Matricula.csv')
dfHorario.to_csv('Horario.csv')

```
